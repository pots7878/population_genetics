```python
#import external modules 
%matplotlib inline 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
#initial parameters 
popsize = 100 #population size 
ngenerations = 100 #number of generations
startp = 0.5 #starting probability 
reps = 100 #number of repetitions 
```

```python
#binomial distribution of generation numbers 
z = np.random.binomial(popsize, startp, reps)
```

```python
#create function sim to simulate generations 

def sim(popsize, ngenerations, startp, reps):
    
    #initialize p to starting probability 
    p0 = startp
    
    #define output vector that defines all alleles 
    alleles = [np.full((reps), int(p0*popsize))]#creates a 2d vector 
    
    #iterate through the number of desired generations 
    for j in range(ngenerations):#list of 0 to ngenerations
        
        #generate the next binomial distribution of popsize 
        x = np.random.binomial(popsize, p0, reps)
        
        #take the number of x and divide it by the popsize to create probability
        p0 = x/popsize
        
        #add probability to alleles list 
        alleles.append(x)
    
    #ouput number of generations and alleles 
    return range(ngenerations+1), alleles
```

```python
#call sim function to simulate number of generations for given popsize
time, a = sim(popsize, ngenerations, startp, reps)
```

```python
#transforming alleles to be able to plot
x = list(zip(*a))
```

```python
#output plot 
for i in x:
    plt.plot(time, i)
plt.ylim(0, popsize)#everytime a line touches the boundary, the spicies goes extinct 
```
